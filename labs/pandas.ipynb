<h3><center>2. Привести описание датасета.</center></h3>

Датасет содержит данные 303 пациентов, которые склонны к сердечно-сосудистым заболеваниям. Всего было взято 14 показателей: 
- возраст; 
- пол (1 - мужчина, 0 - женщина); 
- "cp" = тип боли в груди (0 - типичная стенокардия, 1 - атипичная стенокардия, 2 - неангинальная боль, 3 - бессимптомное); 
- "trestbps" = кровяное давление в состоянии покоя (в мм рт. ст. при поступлении в больницу);
- "chol" = холестерин сыворотки крови в мг/дл;
- "fbs" = уровень сахара в крови натощак > 120 мг/дл (1 = верно; 0 = ложная повторная);
- "restecg" = ЭКГ в покое (0: норма, 1: наличие аномалии ST-T (инверсии Т-волны и/или повышение или понижение ST > 0,05 мВ), 2: вероятная или определенная гипертрофия левого желудочка по критериям Эстеса);
- "thalach" = максимальная достигнутая частота сердечных сокращений;    
- "exang" = стенокардия, вызванная физической нагрузкой (1 = да, 0 = нет);      
- "oldpeak" = депрессия ST, вызванная физической нагрузкой по сравнению с покоем.
- "slope" = наклон пикового сегмента ST при физической нагрузке (0: восходящий, 1: плоский, 2: нисходящий);      
- "ca" = количество крупных сосудов (0-3), окрашенных при флоуроскопии;        
- "thal" = 0 = ошибка (в исходном наборе данных 0 соответствует NaN), 1 = фиксированный дефект, 2 = норма 3 = обратимый дефект;
- "target" = (0 -нет заболевания, 1 = заболевание).

<h4><center>Создание нового поля, вычисленного на основе значений других полей:</center></h4>

- через выражение на базе имеющихся колонок;
- через DataFrame.apply;
- через Series.apply.

*Я подумала, что нужно сделать что-то логичное. У меня такие данные, что просто друг с другом не проссумируешь значения из колонок (потому что результат будет белеберда), поэтому я создала столбец "Риск гипертонии", который будет заполняться, если у пациента высокий холестин и давление.*

- "trestbps" = кровяное давление в состоянии покоя (в мм рт. ст. при поступлении в больницу);
- "chol" = холестерин сыворотки крови в мг/дл;

```python
print(df['trestbps'].unique())
print(df['chol'].unique())
```

*Возьмем пороговые значения давления = 150 и холестерина = 200 (измерение проводится в  мг/дл, а не как в РФ положено ммоль/л!)*

```python
#Создание нового поля, вычисленного на основе значений других полей через выражение на базе имеющихся колонок
df_copy1 = df.copy() #Создаю копию, чтобы не портить датасет 
df_copy1['Risk of hypertension'] = [(
    'yes' if (chol > 200 and trestbps > 150) else 'no'
) for chol, trestbps in zip(df['chol'], df['trestbps'])]

print(df_copy1)
```

```python
#Создание нового поля, вычисленного на основе значений других полей через DataFrame.apply

df_copy2 = df.copy()

def Roh(row): 
    chol = row['chol']
    trestbps = row['trestbps']
    
    if chol > 200 and trestbps > 150:
        return 'yes'
    else:
        return 'no'

df_copy2['Risk of hypertension'] = df_copy2.apply(Roh, axis=1)

print(df_copy2)
```

```python
#Создание нового поля, вычисленного на основе значений других полей через Series.apply.

df_copy3 = df.copy()

def Roh2(chol, trestbps):
    if chol > 200 and trestbps > 150:
        return 'yes'
    else:
        return 'no'

df_copy3['Risk of hypertension'] = df.apply(lambda row: Roh2(row['chol'], row['trestbps']), axis=1)

print(df_copy3)
```

```python
#Группируем по столбцам 'chol', 'cp' и смотрим минимальное и среднее значение давления 
df_grouped = df.groupby(['chol', 'cp'])['trestbps'].agg(['mean', 'min'])
```
